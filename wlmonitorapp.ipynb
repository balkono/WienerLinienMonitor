# A small WienerLinien Monitor by Moritz Emersberger
reading in DIVA(identifier for stations in Vienna) from csv, 
getting realtime data from WienerLinien, 
selecting station and printing all lines departures available

```python

from wlmonitorfuncs import get_data,get_name,get_options,get_time,sel_entry
import csv, os

divalist=[]
with open("divalist.csv", "r") as f:
    csvReader=csv.reader(f, delimiter=",")
    for element in csvReader:
        divalist.insert(0,element)
datalist= get_data(divalist[0])

while True:
    get_options(datalist)
    id=sel_entry()
    if id=="q":
        break
    else:
        id=int(id)
    print("\n"+get_name(datalist[id]))
    for locations in datalist[id]["data"]["monitors"]:
        print("\t"+locations["lines"][0]["name"]+" - "+locations["lines"][0]["towards"])    
        for departures in locations["lines"][0]["departures"]["departure"]:
            print("\t\tin "+str(departures["departureTime"]["countdown"])+" min. at "+get_time(departures["departureTime"]))
        print("")
```
